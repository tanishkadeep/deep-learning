```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sigmoid():
    x = np.linspace(-10, 10, 100)
    y = 1 / (1 + np.exp(-x))

    plt.plot(x, y)
    plt.xlabel("Input")
    plt.ylabel("Sigmoid Output")
    plt.title("Sigmoid Activation Function")
    plt.grid(True)
    plt.show()
```

```python
plot_sigmoid()
```

```python
def plot_relu():
  x=np.linspace(-10,10,100)
  relu=np.maximum(0,x)
  plt.plot(x,relu)
  plt.xlabel("x")
  plt.ylabel("ReLU(x)")
  plt.title("ReLU Activation Function")
  plt.grid(True)
  plt.show()
```

```python
 plot_relu()
```

```python
def plot_leaky_relu():
  x=np.linspace(-10,10,100)
  def leaky_relu(x,alpha=0.01):
    return np.where(x>=0,x,alpha*x)
  leaky_relu_values=leaky_relu(x)
  plt.plot(x,leaky_relu_values)
  plt.xlabel("x")
  plt.ylabel("Leaky ReLU(x)")
  plt.title("Leaky ReLU Activation Function")
  plt.grid(True)
  plt.show()
```

```python
plot_leaky_relu()
```

```python
def plot_tanh():
  x=np.linspace(-10,10,100)
  tanh=np.tanh(x)
  plt.plot(x,tanh)
  plt.xlabel("x")
  plt.ylabel("Tanh(x)")
  plt.title("Hyperbolic Tangent(tanh) Activation Function")
  plt.grid(True)
  plt.show()

plot_tanh()
```

```python
def softmax():
  def softmax_act(x):
    e_x=np.exp(x-np.max(x))
    return e_x/np.sum(e_x,axis=0)
 #Example Usage
  x=np.array([0.25,1.23,-0.0])
  result=softmax_act(x)
  print(result)
  def plot_softmax(probabilities,class_labels):
    plt.bar(class_labels,probabilities)
    plt.xlabel("Class")
    plt.ylabel("Probabilities")
    plt.title("Softmax Output")
    plt.show()
  class_labels=["Class A","Class B","Class C"]
  plot_softmax(result,class_labels)

softmax()
```
