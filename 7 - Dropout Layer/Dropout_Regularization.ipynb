```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('sonar.csv',header=None)
```

```python
data[:]
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.shape[1]-1
```

```python
data[data.shape[1]-1].value_counts()
X = data.drop(data.shape[1]-1,axis=1)
y = data[data.shape[1]-1]
X.head()
```

```python
y = pd.get_dummies(y , drop_first=True)
y.value_counts()
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=1)

X_train.head()
```

```python
import tensorflow as tf
from tensorflow import keras
model = keras.Sequential()
model.add(keras.layers.Dense(60,input_dim=60,activation="relu"))
model.add(keras.layers.Dense(45,activation="relu"))
model.add(keras.layers.Dense(20,activation="relu"))
model.add(keras.layers.Dense(10,activation="relu"))
model.add(keras.layers.Dense(1,activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics = ['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=100,batch_size=32)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test, y_pred))
```

```python
model2 = keras.Sequential()
model2.add(keras.layers.Dense(60,input_dim=60,activation="relu"))
model2.add(keras.layers.Dropout(0.25))
model2.add(keras.layers.Dense(45,activation="relu"))
model2.add(keras.layers.Dropout(0.25))
model2.add(keras.layers.Dense(20,activation="relu"))
model2.add(keras.layers.Dropout(0.25))
model2.add(keras.layers.Dense(10,activation="relu"))
model2.add(keras.layers.Dense(1,activation="sigmoid"))
```

```python
model2.compile(loss="binary_crossentropy" , optimizer = "adam" , metrics = ['accuracy'])
```

```python
model2.fit(X_train,y_train,epochs=100,batch_size=32)
```

```python
model2.evaluate(X_test,y_test)
```

```python
y_pred_2 = model2.predict(X_test).reshape(-1)
y_pred_2 = np.round(y_pred_2)
y_pred_2[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test, y_pred_2))
```
